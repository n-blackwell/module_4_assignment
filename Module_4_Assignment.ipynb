```python
import pandas as pd
import numpy as np
from pprint import pprint
import matplotlib.pyplot as plt
from yellowbrick.cluster import KElbowVisualizer
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn import preprocessing
import scipy.spatial.distance as sp
```

```python
df = pd.read_csv('spotify.csv')
df.fillna(0, inplace=True)
df.set_index(pd.Series(df['song_id']), inplace=True)
df.drop(['song_id'], axis=1, inplace=True)
df
```

```python
temp_audio = df.drop(['uri', 'rank', 'artist_names', 'artists_num', 'artist_individual', 'artist_id', 'artist_genre', 'artist_img', 'collab', 'track_name', 'release_date', 'country', 'region', 'language', 'pivot', 'album_num_tracks', 'album_cover', 'source', 'peak_rank', 'previous_rank', 'weeks_on_chart', 'streams', 'week'], axis=1)
scaler = preprocessing.StandardScaler()
audio = pd.DataFrame(scaler.fit_transform(temp_audio), columns=temp_audio.columns, index=temp_audio.index)
for c in audio.columns:
    audio_mean = audio[c].mean()
    audio[c].fillna(value=audio_mean, inplace=True)
    pd.to_numeric(audio[c])
# pca = PCA(2)
# audio_decomp = pca.fit_transform(audio)
# decomp_df = pd.DataFrame(audio_decomp, index=audio.index)
audio.to_csv('spotify_top_100_audio.csv')
```

```python
model = KMeans()
visualizer = KElbowVisualizer(model, k=(2,50), distance_metric='euclidean')
visualizer.fit(audio)
visualizer.show()
```

```python
kmeans = KMeans(n_clusters=14, random_state=3425)
kmeans.fit_predict(audio)
kmeans.labels_
```

```python
unscaled_audio = scaler.inverse_transform(audio)
audio = pd.DataFrame(unscaled_audio, columns=temp_audio.columns, index=temp_audio.index)
```

```python
cluster_audio = pd.DataFrame(zip(audio.index, kmeans.labels_), columns=["song_id", "cluster"])
cluster_audio.set_index('song_id', inplace=True)
```

```python
cluster_audio["cluster"].value_counts()
```
